```python
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis
X =df[['balance', 'income', 'student2']]
y = df.default2
lda = LinearDiscriminantAnalysis(solver='svd')
y_pred = lda.fit(X, y).predict(X)
df_def = pd.DataFrame({'True default status': y, 'Predicted default status':y_pred})
```

```python
df_def.replace(to_replace={0:'No', 1:'Yes'}, inplace=True)
df_def.groupby(['Predicted default status','True default status']).size().unstack('True default status')
```

```python
decision_prob = 0.2
y_prob = lda.fit(X, y).predict_proba(X)
df_def = pd.DataFrame({'True default status': y,'Predicted default status': y_prob[:,1] > decision_prob})
df_def.replace(to_replace={0:'No', 1:'Yes', 'True':'Yes', 'False':'No'}, inplace=True)
df_def.groupby(['Predicted default status','True default status']).size().unstack('True default status')
```
